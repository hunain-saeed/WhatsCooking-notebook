```python
import json
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def readfile():
        df = pd.read_json('data/train.json')
        return df
```

```python
df = readfile()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['cuisine'].value_counts()
```

```python
df['cuisine'].unique()
```

```python
plt.figure(figsize=(14,7))
graph = df['cuisine'].value_counts().plot(kind='bar', rot=90, fontsize=15)


graph.set_xlabel('Ingredients', fontsize=20, color='navy')
graph.set_ylabel('Number of recipes', fontsize=20, color='navy')
graph.set_title('Total Number of recipes per cuisine', fontsize=25, color='coral')
```

```python
plt.hist(df['ingredients'].str.len(),bins=max(df['ingredients'].str.len()),edgecolor='b')
plt.gcf().set_size_inches(16,8)
plt.title('Ingredients in a Dish Distribution')
```

```python
sns.countplot(y='cuisine', data=df,palette=sns.color_palette('inferno',15))
plt.gcf().set_size_inches(15,10)
plt.title('Cuisine Distribution',size=20)
```
